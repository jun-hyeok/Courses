```
import requests
import csv
import shutil
from bs4 import BeautifulSoup
from datetime import date
```

```
def extract_data(soup):
    titles=soup.select("h4.css-2fgx4k")
    sections=soup.select("p.css-myxawk")
    links=soup.select(".css-e1lvw9 > a")
    data=[]
    for title, section, link in zip(titles, sections, links):
        article=URL+link["href"]
        text=extract_text(article)
        if text!="":
          info={"title":title.text,"date":date,"text":text, "section":section.text,"link":article}
          data.append(info)
    return data
```

```
def extract_text(article):
    html=requests.get(article)
    soup=BeautifulSoup(html.text,"html.parser")
    paragraphs=soup.select("p.css-nnwssh.evys1bk0")
    text=[]
    for paragraph in paragraphs:
        text.append(paragraph.text)
    text="".join(text)
    return text
```

```
def save(articles):
    filename=f"articles_search_{search}.csv"
    file=open(filename, mode="w")
    writer=csv.writer(file)
    writer.writerow(["title","date", "text", "section", "link"])
    for article in articles:
        writer.writerow(list(article.values()))
    shutil.move("/content/"+filename,"/content/drive/My Drive/Colab Notebooks/"+filename)
    return
```
